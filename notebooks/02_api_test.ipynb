```python
import io
import requests
import pandas as pd 

BASE_URL = "http://127.0.0.1:8000"
```

```python
def load_data_parquet(platte: int, nut :float, **params) -> pd.DataFrame:
    params = {"platte": platte, "nut": nut, **params}
    r = requests.get(f"{BASE_URL}/data.parquet", params = params, timeout =120)
    r.raise_for_status()
    return pd.read_parquet(io.BytesIO(r.content))
```

```python
df = load_data_parquet(26,22,limit=80000, wcs_min=-20, wcs_max=80)
df.shape, df[["WCS_Y_mm"]].describe()
```
